```python
import pandas as pd

df = pd.read_csv("../vehicles_us.csv")
df.head()
```

```python
df.info()
```

```python
df['model_year'] = df['model_year'].fillna(df['model_year'].mean()).astype(int)
df['model'] = df['model'].str.upper()
df['condition'] = df['condition'].str.upper().fillna('UNKNOWN').astype('category')
df['cylinders'] = df['cylinders'].fillna(df['cylinders'].mean()).astype(int)
df['fuel'] = df['fuel'].str.upper().astype('category')
df['odometer'] = df['odometer'].fillna(df['odometer'].mean()).astype(int)
df['transmission'] = df['transmission'].str.upper()
df['type'] = df['type'].str.upper()
df['paint_color'] = df['paint_color'].str.upper().fillna('UNKNOWN').astype('category')
df['date_posted'] = pd.to_datetime(df['date_posted'])
df['is_4wd'] = df['is_4wd'].fillna('0').astype('int')
df.columns = df.columns.str.upper()
```
